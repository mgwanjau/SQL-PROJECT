```python
%load_ext sql
%sql sqlite://
import pandas as pd
accidents = pd.read_csv('traffic_accidents.csv')
%sql DROP TABLE if EXISTS accidents;
%sql PERSIST accidents;
%sql SELECT * FROM accidents LIMIT 5;
```

```python
%%sql
SELECT MAKE, COUNT(MAKE) 
FROM accidents 
GROUP BY MAKE 
ORDER BY COUNT(MAKE) DESC LIMIT 5;
```

```python
%%sql
SELECT TRAVEL_DIRECTION, VEHICLE_USE, MAX(VEHICLE_USE_count)
FROM (SELECT TRAVEL_DIRECTION, VEHICLE_USE, COUNT(VEHICLE_USE)  AS VEHICLE_USE_count
FROM accidents 
GROUP BY TRAVEL_DIRECTION, VEHICLE_USE
ORDER BY TRAVEL_DIRECTION, VEHICLE_USE_count DESC)
GROUP BY TRAVEL_DIRECTION;
```

```python
%%sql
SELECT COUNT(NUM_PASSENGERS) 
FROM accidents 
WHERE VEHICLE_USE ="PERSONAL" 
```

```python
%%sql
SELECT COUNT(NUM_PASSENGERS) 
FROM accidents 
WHERE VEHICLE_TYPE ="SPORT UTILITY VEHICLE (SUV)" 
```

```python
%%sql
SELECT COUNT(*) FROM accidents WHERE MAKE="MERCEDES-BENZ" AND MANEUVER="PARKED";
```

```python
%%sql
SELECT TRAVEL_DIRECTION, MAX(NUM_PASSENGERS_count)
FROM (SELECT TRAVEL_DIRECTION,COUNT(NUM_PASSENGERS) AS NUM_PASSENGERS_count
FROM accidents 
GROUP BY TRAVEL_DIRECTION
ORDER BY NUM_PASSENGERS_count ASC)
GROUP BY TRAVEL_DIRECTION;
```
